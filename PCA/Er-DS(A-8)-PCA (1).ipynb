# Heirarchical Clustering

```python
hc_df=pca_df.copy()
```

```python
# create dendrogram
plt.figure(figsize=(20,10))
dendrogram = sch.dendrogram(sch.linkage(hc_df, method='complete'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
```

```python
hc.fit(hc_df)
```

```python
# save clusters for chart
y_hc = hc.fit_predict(hc_df)
```

```python
y_hc
```

```python
hc_df['cluster_id'] = y_hc+1
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=hc_df[0], y=hc_df[1], hue=hc_df.cluster_id.tolist(),
                 palette=sns.color_palette("bright", 3),
                data=hc_df).set(title="Heirarchical projection") 
plt.show()
```

# K-Means Clustering

```python
km_df=pca_df.copy()
```

```python
plt.figure(figsize=(10,7))
wcss2 = []
for i in range(1,11):
  kmeans2 = KMeans(n_clusters=i,random_state=0)
  kmeans2.fit(km_df)
  wcss2.append(kmeans2.inertia_)

plt.style.use("fivethirtyeight")
plt.plot(range(1,11),wcss2, marker = "*")
plt.title("Elbow Method")
plt.xlabel("No. of Clusters")
plt.ylabel("WCSS")
plt.show()
```

>Looking at the curve we may conclude that the number of clusters shall be taken as 3.

## Checking Elbow point :

```python
kl = KneeLocator(range(1, 11), wcss2, curve="convex", direction="decreasing")
kl.elbow
```

* The scree plot levels off at k=3 and let's use it to determine the clusters.

```python
#Build Cluster algorithm

clusters_new = KMeans(3, random_state=42)
clusters_new.fit(km_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
km_df['kmeans_cluster_id'] = clusters_new.labels_
```

```python
km_df.groupby('kmeans_cluster_id').agg(['mean']).reset_index()
```

```python
km_df.head()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=km_df[0], y=km_df[1], hue=km_df.kmeans_cluster_id.tolist(),
                 palette=sns.color_palette("bright", 3),
                data=km_df).set(title="K-Means projection") 
plt.show()
```
